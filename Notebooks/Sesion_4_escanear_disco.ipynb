## ¿Qué es una ruta?

Una **ruta** es la dirección de un archivo dentro del disco.
Es como la dirección postal de un archivo.

```
E:\NUEVO ORIENTE ESTEREO\PROGRAMAS\BARRIO ADENTRO\BARRIO ADENTRO 06-06-25.mp3
│                         │          │              │
disco                     carpeta    subcarpeta     archivo
```

**El problema:** Windows usa `\` para separar carpetas.
Linux y Mac usan `/`. Si escribimos rutas como texto puro,
el código falla en cualquier sistema diferente al que lo escribimos.

`pathlib` resuelve esto automáticamente.

## Parte 4: `try/except`: manejar errores sin romper el sistema

El disco de Oriente Estéreo tiene 2.548 archivos. Garantizado
que alguno va a tener un problema: archivo dañado, nombre raro,
permisos bloqueados, disco desconectado a mitad del escaneo.

Sin manejo de errores, el sistema se rompe en el archivo 47
y pierde los otros 2.501.

Con `try/except`, registra el problema y continúa.

> 💡 **Analogía:** el locutor va a poner una canción y el archivo
> está dañado. No se va del estudio, anota el problema
> y pone la siguiente canción.

```python
try:
    # código que puede fallar
except TipoDeError:
    # qué hacer si falla — sin romper el programa
```


````python
# Sin try/except: el primer error para todo
# Con try/except: registra el error y continúa

archivos = [
    Path("/content/disco_prueba/PROGRAMAS/BARRIO ADENTRO/BARRIO ADENTRO 06-06-25.mp3"),
    Path("/content/disco_prueba/PROGRAMAS/ARCHIVO_QUE_NO_EXISTE.mp3"),  # ← no existe
    Path("/content/disco_prueba/PROGRAMAS/EN LA RAYA/EN LA RAYA DEPORTES DIC 1 DE 2025.mp3"),
]

print("Con try/except — el sistema continúa aunque haya errores:")
for arch in archivos:
    try:
        stats = arch.stat()   # falla si el archivo no existe
        print(f"  ✅ {arch.name}: {stats.st_size} bytes")
    except FileNotFoundError:
        print(f"  ❌ No encontrado: {arch.name}")
    except PermissionError:
        print(f"  🔒 Sin permisos : {arch.name}")
    except Exception as error:
        print(f"  ⚠️  Error inesperado en {arch.name}: {error}")

print()
print("El sistema procesó todos los archivos posibles ✅")
````
